```python
mov=df.dropna()
```

```python
mov.isnull().sum()
```

```python
mov= mov[['id','title','overview','genre']]
mov
```

```python
mov['tags']= mov['overview']+mov['genre']
mov
```

```python
mov= mov.drop(columns=['overview','genre'])
mov
```

```python
mov=mov.reset_index()
mov
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=9985, stop_words='english')
```

```python
cv
```

```python
vector= cv.fit_transform(mov['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

# Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
mov[mov['title']=="Domino"].index[0]
```

```python
mov.info()
```

```python
similarity[2]
```

```python
distance=sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])

for i in distance[0:5]:
  print(mov.iloc[i[0]].title)
```

```python
def recommend(movies):
  index=mov[mov['title']==movies].index[0]
  distance= sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
  for i in distance[0:5]:
    print(mov.iloc[i[0]].title)
```

```python
recommend("It")
```
